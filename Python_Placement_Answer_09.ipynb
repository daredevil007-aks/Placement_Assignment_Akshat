```python
import pandas as pd #importing the required library
import matplotlib.pyplot as plt
import seaborn as sns
import squarify

df = pd.read_csv(r'C:\Users\hp\Downloads\Electric_Vehicle_Population_Data.csv')
df.head()
```

```python
df = df.dropna() #deleting the columns which carries NA values 
df = df.drop(["VIN (1-10)"],axis =1) #dropping the unwanted column
```

```python
not_qualified = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Not eligible due to low battery range']
not_qualified[['Model','ElectricVehicleType']] 
#getting the Model & ElectricVehicleType which are Not eligible due to low battery range
```

```python
tesla_car = (df.loc[(df['City']=='Bothell')& (df['Make']=='TESLA'),['Model Year','Model']]) #applying different conditions at a time
tesla_car
```

```python
car = (df.loc[(df['Electric Range']> 100)& (df['Model Year'] > 2015)])
car
```

```python
counts = df.groupby(['City', 'Electric Vehicle Type']).size().unstack() #distribution of electric vehicle types

# Aggregate the data by EV Type
agg_counts = counts.sum()

# Plotting a grouped bar plot with aggregated data
agg_counts.plot(kind='bar')
plt.xlabel('EV Type')
plt.ylabel('City')
plt.title('Distribution of Electric Vehicle Types')
plt.show()
```
