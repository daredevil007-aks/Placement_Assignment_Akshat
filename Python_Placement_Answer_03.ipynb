```python
import pandas as pd
import json
import urllib.request

url = 'https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json'

# Download the JSON file
filename, headers = urllib.request.urlretrieve(url, filename="pokedex.json")

# Read the JSON file into a Pandas DataFrame
with open(filename) as json_file:
    data = json.load(json_file)
    df = pd.DataFrame(data['pokemon'])
df.head(10)
```

```python
df.shape() #getting how many rows and coulmns are their in dataset.
```

```python
df.describe(include='all')
```

```python
data_copy = df.copy() #making copy of a dataset
```

```python
data_copy ['spawn_time'] == 'N/A'
```

```python
data_copy['spawn_time'] =data_copy['spawn_time'].str.replace('N/A','00:00')
```

```python
data_copy['spawn_time'] = data_copy['spawn_time'].apply(lambda x: '00:' + str(x)) #using lambda function to add into hours format.
```

```python
data_copy['spawn_time']
```

```python
data_copy['spawn_time'] = pd.to_timedelta(data_copy['spawn_time'])  #datytime conversion
```

```python
data_copy['spawn_time']
```

```python
rt = data_copy.copy()
```

```python
rt.index = rt.index +1 #reseting the index values 
rt['next_evolutio'] = rt.apply(lambda row: [{'num': str(row['num'][0:]), 'name': row['name'][0:]}], axis=1)  #making new column which takes name and number into dict format with the help of lambda function 
rt['next_evolutio'] = rt['next_evolutio'].reset_index(drop=True)
rt
```

```python
rt['next_evolution'] = rt['next_evolution'].fillna(rt['next_evolutio']) #fill new column values where the cell is empty
rt
```

```python
rt.index = rt.index -1
rt['prev_evolutio'] = rt.apply(lambda row: [{'num': str(row['num'][0:]), 'name': row['name'][0:]}], axis=1) #making new column which takes name and number into dict format with the help of lambda function 
rt['prev_evolutio'] = rt['prev_evolutio'].reset_index(drop=True)
 
rt
```

```python
rt['prev_evolution'] = rt['prev_evolution'].fillna(rt['prev_evolutio']) #fill new column values where the cell is empty in a coulmn row
rt
```

```python
rt.drop(['next_evolutio','prev_evolutio'], axis=1) #droping the unwante column
```

```python
rt.to_csv('final.csv') #saving it to CSV file
```
