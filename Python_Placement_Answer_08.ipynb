```python
import pandas as pd
import plotly as pt
import seaborn as sns
import matplotlib.pyplot as plt
import csv
import plotly.express as px

paths =r'ty.csv'

df = pd.read_csv(paths)
df
```

```python
average_ratings = df.groupby('season')['rating.average'].mean()  #taking mean value of every season and plotting it on graph
average_ratings
average_ratings.plot(kind='bar')
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.title('Average Rating per Season')
plt.show()
```

```python
average_ratings = df.groupby('season')['rating.average'].mean() #using groupby function to groupby all the seasons
average_ratings > 8 #only season 1 is having average rating of above 8.
```

```python
gt = df['airdate'] < ('2019-05-01 09:00:00') #filtering episodes before 2019-may.
true_values = df[gt == True]
true_values
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='name', data=true_values)
plt.xlabel('Count')
plt.ylabel('Episode Name')
plt.title('Episode Names Aired Before May 2019')
plt.show()
```

```python

#using lambda function to get the maximum and minimum of each season separately
highest_rating_episodes = df.groupby('season').apply(lambda x: x.loc[x['rating.average'].idxmax()]) 
lowest_rating_episodes = df.groupby('season').apply(lambda x: x.loc[x['rating.average'].idxmin()])

# Get the episode name, highest rating, and lowest rating for each season
result = pd.DataFrame({
    'Season': highest_rating_episodes['season'],
    'Highest Rated Episode': highest_rating_episodes['name'],
    'Highest Rating': highest_rating_episodes['rating.average'],
    'Lowest Rated Episode': lowest_rating_episodes['name'],
    'Lowest Rating': lowest_rating_episodes['rating.average']
})

# Print the result
print(result)
```

```python
plt.figure(figsize=(10, 6))

# Plot the highest and lowest ratings as point plots
sns.pointplot(x='Season', y='Highest Rating', data=result, color='green')
sns.pointplot(x='Season', y='Lowest Rating', data=result, color='red')

# Customize the plot
plt.xlabel('Season')
plt.ylabel('Rating')
plt.title('Episode Names with Highest and Lowest Ratings per Season')
plt.legend()

# Show the plot
plt.show()
```
