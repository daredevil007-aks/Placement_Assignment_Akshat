```python
import requests
import pandas as pd

url = 'http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes'
response = requests.get(url) #GET request to the API to get the data

data = response.json() 
data
```

```python
embedded_data = data['_embedded'] #retriving data from the JSON file
```

```python
final = embedded_data['episodes'] 
```

```python
df = pd.json_normalize(final, max_level=1) 
df.head(2)
```

```python
df['id'] = df['id'].astype(int)  #type conversion of columns in integer, string, float.
```

```python
df['url'] = df['url'].astype(str)
```

```python
df['name'] = df['name'].astype(str)
```

```python
df['season'] = df['season'].astype(int)
```

```python
df['number'] = df['number'].astype(int)
```

```python
df['type'] = df['type'].astype(str)
```

```python
df['airdate'] =pd.to_datetime(df['airdate'])
```

```python
df['airtime'] = pd.to_datetime(df['airtime'])
df['airtime'] = df['airtime'].dt.strftime('%I:%M %p')
```

```python
df['runtime'] = df['runtime'].astype(float)
```

```python
df['rating.average'] = df['rating.average'].astype(float)
```

```python
df['summary'] = df['summary'].astype(str)
```

```python
df['image.medium'] = df['image.medium'].astype(str)
```

```python
df['image.original'] = df['image.original'].astype(str)
```

```python
df['summary'] = df['summary'].str.replace('<p>','')
```

```python
df['summary'] = df['summary'].str.replace('</p>','')
```

```python
df.to_csv('ty.csv') #saving in CSV format.
```
