```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv(r"C:\Users\hp\Desktop\data.csv")
```

```python
X = df.drop(['label','filename'], axis =1).values
y = df['label'].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
km = KMeans(n_clusters=5, random_state=42)
km.fit(X_scaled)
```

```python
cluster_labels = km.labels_
cluster_labels
```

```python
silhouette_avg = silhouette_score(X_scaled, cluster_labels)
print("Score:", silhouette_avg)
```

```python
km = KMeans(n_clusters=5, random_state=42)
km_labels = km.fit_predict(X_scaled)
km_score = silhouette_score(X_scaled, km_labels)
```

```python
agg = AgglomerativeClustering(n_clusters=5)
agg_label = agg.fit_predict(X_scaled)
agg_score = silhouette_score(X_scaled, agg_label)
```

```python
gmm = GaussianMixture(n_components=5, random_state=42)
gmm.fit(X_scaled)
gmm_labels = gmm.predict(X_scaled)
gmm_score= silhouette_score(X_scaled, gmm_labels)
```

```python
print("k-mean score:", km_score)
print("Agglomerative score:", agg_score)
print("GMM Score:" ,gmm_score)
```
