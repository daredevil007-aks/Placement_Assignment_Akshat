```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import re
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r"C:\Users\hp\Desktop\ObesityDataSet_raw_and_data_sinthetic.csv")
data.head()
```

```python
figure = px.scatter(data, x="Weight",
                   y = "Height",trendline = "ols", color="Age",
                   title = "Relationship Between Weight and Height by Age")
figure.show()
```

```python
data.isna().sum()
```

```python
encoder = LabelEncoder()
data['Gender'] = encoder.fit_transform(data['Gender'])
data['family_history_with_overweight'] = encoder.fit_transform(data['family_history_with_overweight'])
data['FAVC'] = encoder.fit_transform(data['FAVC'])
data['CAEC'] = encoder.fit_transform(data['CAEC'])
data['SMOKE'] = encoder.fit_transform(data['SMOKE'])
data['SCC'] = encoder.fit_transform(data['SCC'])
data['CALC'] = encoder.fit_transform(data['CALC'])
data['MTRANS'] = encoder.fit_transform(data['MTRANS'])
data['NObeyesdad'] = encoder.fit_transform(data['NObeyesdad'])
```

```python
X = data.drop('NObeyesdad', axis =1)
Y = data['NObeyesdad']
```

```python
X_Train, X_Test, y_Train, y_Test = train_test_split(X, Y, test_size=0.2, random_state=42)
logreg_model = LogisticRegression()
logreg_model.fit(X_Train, y_Train)
logreg_predict = logreg_model.predict(X_Test)
logreg_accuracy = accuracy_score(y_Test, logreg_predict)
print("Logistic Regression Accuracy:", logreg_accuracy)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_Train, y_Train)
rf_predict = rf_model.predict(X_Test)
rf_accuracy = accuracy_score(y_Test, rf_predict)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
svm_model = SVC()
svm_model.fit(X_Train, y_Train)
svm_predict = svm_model.predict(X_Test)
svm_accuracy = accuracy_score(y_Test,svm_predict)
print("SVM Accuracy:", svm_accuracy)
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_Train,y_Train)
knn_predict = knn_model.predict(X_Test)
knn_accuracy = accuracy_score(y_Test, knn_predict)
print("KNN Model Accuracy:",knn_accuracy)
```
