```python
import pandas as pd
import plotly as pt
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('final.csv')
data.head(2)
```

```python
data['new'] = data['avg_spawns'] < 5  
true_values = data[data['new'] == True]
true_values.head(2)
```

```python
data['avg_spawns'] = data['avg_spawns'] < 5  #ploting the avg_spawns less than 5 in the graph
sns.countplot(data=data, x='avg_spawns', color='purple')
plt.xlabel('Average Spawns < 5')
plt.ylabel('Count')
plt.title('Distribution of Average Spawns')
plt.show()
```

```python
data['Length_weakness'] = data.weaknesses.apply(lambda x: len(x))
data.head()
```

```python
data_pu = data['multipliers'].isnull().value_counts() #using seaborn to plo multipliers with no values
sns.countplot(data=data, x=data_pu, color='purple')
plt.title('multiplers with no values ')
plt.xticks(rotation=90);
```

```python
data['next_evolution'] =data['next_evolution'].tolist()
```

```python
import numpy as np
data['List_Length'] = data['next_evolution'].apply(lambda x: len(x) if isinstance(x, list) else np.nan)
```

```python
data['spawn_time'] = pd.to_timedelta(data['spawn_time'])

filtered_pokemon = data[data['spawn_time'] < pd.Timedelta(minutes=5)]  #appyling condition where time is not more than 5 mins using pandas Timedelta function
filtered_pokemon
```
