```python
import pandas as pd
import seaborn as sns
import plotly as plt
df = pd.read_csv('metero.csv', parse_dates=['year'])
df
```

```python
df['below_2000'] = df['year'] < '2000'   #taking below_2000 as new column which holds timeframe below 2000
true_values = df[df['below_2000'] == True]
true_values.name
sns.countplot(data=df, x='below_2000', color='purple')  #plotting it with seaborn
```

```python
df['below_1970'] = df['year'] < '1970'    #taking below_1970 as new column which holds timeframe below 1970
true_value = df[df['below_1970'] == True]
true_value
sns.countplot(data=df, x='below_1970', color='purple') #plotting it with seaborn
```

```python
df['mass_over'] = df['mass'] > 10000   #taking mass_over as new column which holds mass of above 10000
true_value = df[df['mass_over'] == True]
true_value.mass
```
